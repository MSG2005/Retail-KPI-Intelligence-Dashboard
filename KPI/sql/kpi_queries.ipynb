```python
# ===============================
# 1. Connect to SQLite Database
# ===============================

import sqlite3
import pandas as pd

conn = sqlite3.connect("superstore.db")

cursor = conn.cursor()
```

```python
# ===============================
# 2. Load Cleaned CSV into SQL
# ===============================

df = pd.read_csv("superstore_cleaned.csv")

df.to_sql("sales", conn, if_exists="replace", index=False)

print("Data successfully loaded into SQL database.")
```

```python
#Total Revenue

query = """
SELECT SUM(Sales) AS Total_Revenue
FROM sales;
"""

pd.read_sql(query, conn)
```

```python
#Profit Margin %

query = """
SELECT 
    (SUM(Profit) * 100.0 / SUM(Sales)) AS Profit_Margin
FROM sales;
"""

pd.read_sql(query, conn)
```

```python
#Revenue by Region

query = """
SELECT Region,
       SUM(Sales) AS Revenue
FROM sales
GROUP BY Region
ORDER BY Revenue DESC;
"""

pd.read_sql(query, conn)
```

```python
#Churn Rate

query = """
SELECT 
    (COUNT(DISTINCT CASE WHEN "Churn Flag" = 1 THEN "Customer ID" END) * 100.0
     / COUNT(DISTINCT "Customer ID")) AS Churn_Rate
FROM sales;
"""

pd.read_sql(query, conn)
```

```python
#Top 10 Customers

query = """
SELECT "Customer Name",
       SUM(Sales) AS Revenue
FROM sales
GROUP BY "Customer Name"
ORDER BY Revenue DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```
